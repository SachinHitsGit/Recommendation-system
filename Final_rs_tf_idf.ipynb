```python
import pandas as pd

ratings = pd.read_csv(
    "ml-1m/ratings.dat",
    sep="::",
    engine="python",
    encoding="latin-1",
    names=["userId", "movieId", "rating", "timestamp"],
    header=None
)

movies = pd.read_csv(
    "ml-1m/movies.dat",
    sep="::",
    engine="python",
    encoding="latin-1",
    names=["movieId", "title", "genres"],
    header=None
)
```

```python
movies.head()
```

```python
ratings.head()
```

```python
movies["content"] = movies["title"] + " " + movies["genres"].str.replace("|", " ")
movies.head()
```

```python
data = ratings.merge(movies, on="movieId")
data.head(20)
```

```python
documents = movies['content'].fillna("").tolist()
print(type(documents))
print(documents[:10])
```

```python
tokenized_docs = []

for doc in documents:
    doc = doc.lower()
    doc = doc.replace("(","").replace(")","")
    doc = doc.replace(",","").replace(".","")
    doc = doc.replace("'","")
    doc = doc.split()
                  
                        
                                       
    tokenized_docs.append(doc)

print(tokenized_docs[:20])
print(len(tokenized_docs))
```

```python
vocab = set()
for doc in tokenized_docs:
    vocab.update(doc)
print(vocab)
print(len(vocab))
```

```python
vocab = sorted(list(vocab))
print(vocab[:500])
```

Column Index

```python
word_to_index = {}

for i, word in enumerate(vocab):
    word_to_index[word] = i

print(word_to_index["1"])
```

Create empty matrix

```python
import numpy as np

num_docs = len(tokenized_docs)
num_words = len(vocab)

                                              
                                                                                          
                                                

count_matrix = np.zeros((num_docs,num_words),dtype=np.float64)

print(count_matrix.shape)
```

```python
count_matrix[:,0]
```

```python
from collections import Counter

for row_doc_index, doc in enumerate(tokenized_docs):
    word_count_dic = Counter(doc)
    for word in doc:
        col_word_index = word_to_index[word]
        count_matrix[row_doc_index][col_word_index] = word_count_dic[word]
print(count_matrix.shape)
```

```python
import pandas as pd

df_matrix = pd.DataFrame(
    count_matrix,
    columns=list(word_to_index.keys()),                        
    index=[f"Movie_{i}" for i in range(len(tokenized_docs))]                      
)

print(df_matrix)
```

Initilizing tf_matrix

```python
tf_matrix = np.zeros_like(count_matrix)

for i in range(num_docs):
    total_words = np.sum(count_matrix[i])
    if total_words != 0:
        tf_matrix[i] = count_matrix[i] / total_words
print(tf_matrix)
```

Compute df

```python
print(len(count_matrix[6]))
print(count_matrix[5].shape)
```

DF(t) = frequency of each words

```python
df = np.zeros(num_words,)
df.shape
for i in range(num_words):
    df[i] = np.sum(count_matrix[:,i] > 0)
print(df)
```

N = total number of documents (constant)
IDF = log(N/df(t))

```python
N = num_docs
idf = np.log(N/(df+1))
```

```python
print("tf_matrix.shape : ",tf_matrix.shape)
print("idf.shape : ",idf.shape)
tfidf_matrix = tf_matrix * idf
print("tfidf_matrix : ",tfidf_matrix.shape)
```

Done the TF-IDF matrx. 

Now make a user profile to recommend to that user

Get User - 1 's high rated movies

```python
user_id = 1
high_rated = ratings[(ratings['userId'] == user_id) & (ratings['rating'] >= 4)]
high_rated[:5]
```

```python
liked_movies_ids = high_rated['movieId']
liked_indices = movies[movies['movieId'].isin(liked_movies_ids)].index.tolist()
```

```python
movies[movies['movieId'].isin(liked_movies_ids)]["title"]
```

```python
print(liked_indices)

for idx in liked_indices[:3]:
    print("Index:", idx)
    print("Movie title:", movies.iloc[idx]['title'])
    print("Vector shape:", df_matrix.iloc[idx].shape)
    print()
    
```

Build the User Profile Vector

```python
user_profile = np.mean(tfidf_matrix[liked_indices],axis=0)
print(user_profile)
```

Compute Cosine Similarity (User vs All Movies)

```python
user_norm = np.sqrt(np.sum(user_profile ** 2))
movie_norm = np.sqrt(np.sum(tfidf_matrix**2, axis=1))
dot_product = tfidf_matrix @ user_profile

similarities = dot_product / (user_norm * movie_norm)

similarity = np.nan_to_num(similarities)
similarity[1]
```

Remove Movies Already Seen

```python
similarity[liked_indices] = -1
                                  
```

Get Top 10 Recommendations

```python
top_ten_indices = np.argsort(similarity)[::-1][:10]
recommended_movies = movies.iloc[top_ten_indices]["title"]
print(recommended_movies)
```

Mean Recall@10 across many users

```python
import numpy as np

users = ratings["userId"].unique()

recalls = []

for user in users:

    user_data = ratings[ratings["userId"] == user]

    liked_movies = user_data[user_data["rating"] >= 4]["movieId"].tolist()

    if len(liked_movies) < 5:
        continue

    split = int(0.8 * len(liked_movies))

    train_movies = liked_movies[:split]
    test_movies = liked_movies[split:]

    train_indices = movies[movies["movieId"].isin(train_movies)].index.tolist()

    if len(train_indices) == 0:
        continue

    user_profile = np.mean(tfidf_matrix[train_indices], axis=0)

    dot_product = tfidf_matrix @ user_profile

    user_norm = np.sqrt(np.sum(user_profile ** 2))
    movie_norm = np.sqrt(np.sum(tfidf_matrix ** 2, axis=1))

    similarity = dot_product / (user_norm * movie_norm)

    similarity[train_indices] = -1

    top_indices = np.argsort(similarity)[::-1][:10]

    recommended_movie_ids = movies.iloc[top_indices]["movieId"].tolist()

    hits = len(set(recommended_movie_ids) & set(test_movies))

    recall = hits / len(test_movies)

    recalls.append(recall)

average_recall = np.mean(recalls)

print("Average Recall@10:", average_recall)
```

Leave-one-out evaluvation

```python
import numpy as np

users = ratings["userId"].value_counts()
users = users[users > 20].index

recalls = []

movie_norm = np.sqrt(np.sum(tfidf_matrix ** 2, axis=1))

for user in users:

    user_data = ratings[ratings["userId"] == user]

    liked_movies = user_data[user_data["rating"] >= 4]["movieId"].tolist()

    if len(liked_movies) < 5:
        continue

                         
    test_movie = liked_movies[-1]
    train_movies = liked_movies[:-1]

    train_indices = movies[movies["movieId"].isin(train_movies)].index.tolist()

    if len(train_indices) == 0:
        continue

    ratings_weights = user_data[user_data["movieId"].isin(train_movies)]["rating"].values

    user_profile = np.average(tfidf_matrix[train_indices], axis=0, weights=ratings_weights)

    dot_product = tfidf_matrix @ user_profile

    user_norm = np.sqrt(np.sum(user_profile ** 2))

    similarity = dot_product / (user_norm * movie_norm)

    similarity[train_indices] = -1

    top_indices = np.argsort(similarity)[::-1][:10]

    recommended_movie_ids = movies.iloc[top_indices]["movieId"].tolist()

                                          
    if test_movie in recommended_movie_ids:
        recall = 1
    else:
        recall = 0

    recalls.append(recall)

average_recall = np.mean(recalls)

print("Recall@10 (HitRate@10):", average_recall)
```
