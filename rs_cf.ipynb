```python
import pandas as pd

ratings = pd.read_csv(
    "ml-1m/ratings.dat",
    sep="::",
    engine="python",
    encoding="latin-1",
    names=["userId", "movieId", "rating", "timestamp"],
    header=None
)

movies = pd.read_csv(
    "ml-1m/movies.dat",
    sep="::",
    engine="python",
    encoding="latin-1",
    names=["movieId", "title", "genres"],
    header=None
)
```

```python
movies["content"] = movies["title"] + " " + movies["genres"].str.replace("|", " ")
movies.head()
```

```python
ratings
```

```python
data = ratings.merge(movies, on="movieId")
data.head(5)
```

Build the User–Movie Rating Matrix

```python
ratings_matrix = ratings.pivot(index="userId",columns="movieId",values="rating")
ratings_matrix.fillna(0)
```

SVD requires a numeric matrix.

```python
import numpy as np

R = ratings_matrix.to_numpy()
print(R.shape)
```

normalize rating using mean

```python
user_mean = np.mean(R, axis = 1)
R_centered = R - user_mean
```
